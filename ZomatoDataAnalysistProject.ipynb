Zomato Data Analysis Project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

create data frame

```python
dataframe = pd.read_csv('Zomato data .csv')
print(dataframe)
```

```python
dataframe
```

conver the datatype of column - rating

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe
```

```python
dataframe.info()
```

Question 1 : type of resturent

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of resturent")
```

conclusion : major of the resturent fall in dinning category

```python
dataframe.head()
```

Question 2 : votes has received from customers

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="red", marker="*")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

Conclusion : dinning resturent has received maximum votes

Question 6 : which type of resturent received more offline orders,so that zomato can provide customers with some gud offers

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlOrBr", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

conclusion : Dinning resturent primarily accept offline orders,whereas cafes primarily receive online orders.This suggests that clients prefer orders in person at resturents,but prefer online ordering at cafes
